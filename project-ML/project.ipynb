```python
import csv
from collections import defaultdict

disease_list = []

def return_list(disease):
    disease_list = []
    match = disease.replace('^','_').split('_')
    ctr = 1
    for group in match:
        if ctr%2==0:
            disease_list.append(group)
        ctr = ctr + 1

    return disease_list

with open("/home/mudit/practice/Scraped-Data/dataset_uncleaned.csv") as csvfile:
    reader = csv.reader(csvfile)
    disease=""
    weight = 0
    disease_list = []
    dict_wt = {}
    dictt_=defaultdict(list)
    for row in reader:
        if row[0]!="\xc2\xa0" and row[0]!="":
            disease = row[0]
            disease_list = return_list(disease)
            weight = row[1]

        if row[2]!="\xc2\xa0" and row[2]!="":
            symptom_list = return_list(row[2])

            for d in disease_list:
                for s in symptom_list:
                    dictt_[d].append(s)
                dict_wt[d] = weight

with open("/home/mudit/practice/Scraped-Data/dataset_clean.csv","w") as csvfile:
    writer = csv.writer(csvfile)
    for key,values in dict_.items():
        for v in values:
            #key = str.encode(key)
            #key = str.encode(key)
            #.strip()
            v = v.encode('utf-8').strip()
            #v = str.encode(v)
            writer.writerow([key,v,dict_wt[key]])               
                
    #print (dict_)
```

```python
data = pd.read_csv("/home/mudit/practice/Scraped-Data/dataset_clean.csv", encoding ="ISO-8859-1")
```

```python
df = pd.DataFrame(data)
```

```python
df_1 = pd.get_dummies(df.Target)
```

```python
df_s = df['Source']
```

```python
df_pivoted = pd.concat([df_s,df_1], axis=1)
```

```python
df_pivoted.drop_duplicates(keep='first',inplace=True)
print(df_pivoted)
```

```python
cols = df_pivoted.columns
```

```python

cols=cols[1:]
```

```python
df_pivoted = df_pivoted.groupby('Source').sum()
df_pivoted = df_pivoted.reset_index()
df_pivoted[:5]
```

```python
df_pivoted.to_csv("/home/mudit/practice/Scraped-Data/df_pivoted.csv")
```

```python
x = df_pivoted[cols]
y = df_pivoted['Source']
```

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.naive_bayes import MultinomialNB
from sklearn.cross_validation import train_test_split



x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=42)

mnb = MultinomialNB()
mnb = mnb.fit(x_train, y_train)
mnb.score(x_test, y_test)

mnb_tot = MultinomialNB()
mnb_tot = mnb_tot.fit(x, y)

mnb_tot.score(x, y)

disease_pred = mnb_tot.predict(x)

disease_real = y.values

for i in range(0, len(disease_real)):
    if disease_pred[i]!=disease_real[i]:
        print ('Pred: {0} Actual:{1}'.format(disease_pred[i], disease_real[i]))
```
